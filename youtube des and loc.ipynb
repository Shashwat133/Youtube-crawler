```python
class youtubedetails:
  def __init__(self, url):
    self.url = url
  def getdata(self):
    options = Options()
    ua = UserAgent()
    userAgent = ua.random
    options.add_argument(f'user-agent={userAgent}')
    options = webdriver.ChromeOptions()
    preferences= {"profile.default_content_settings.popups": 0,"download.default_directory":r"\data","directory_upgrade": True}
    options.add_experimental_option("prefs",preferences)
    chrome = r'C:\Users\shardul\printPDF\chromedriver_win32\chromedriver.exe'
    #options.add_extension(r'C:\Users\shardul\Downloads\extension_3_9_5_0.crx')
    driver = webdriver.Chrome(chrome,chrome_options = options)
    driver.get(self.url)
    driver.maximize_window()
    wait = WebDriverWait(driver, 10)
    about = wait.until(EC.element_to_be_clickable((By.XPATH,'//*[@id="tabsContent"]/tp-yt-paper-tab[6]')))
   
    about.click()
    try:
        des = wait.until(EC.element_to_be_clickable((By.XPATH,'//*[@id="description"]'))).text
    except:
        des='Not found'
    try:
        loca = wait.until(EC.element_to_be_clickable((By.XPATH,'//*[@id="details-container"]/table/tbody/tr[2]/td[2]/yt-formatted-string'))).text
    except:
        loca='Not found'
    driver.quit()
    return des,loca
```

```python
src_df = pd.read_excel('Book3.xlsx')
```

```python
df = pd.DataFrame(columns = ['Des','Location','Channel Url'])
for i, row in src_df.iterrows():
    print(i)
    url = row['websites']
    p1 = youtubedetails(url)
    dex,loca= p1.getdata()
    print(dex,loca)
    df=df.append({'Des':dex, 'Location':loca,'Channel Url':url},ignore_index = True)

    
```
